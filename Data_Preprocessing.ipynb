# Data Preprocessing Core Functionalities

```python
# from txt to json file 
import json
from normalizer import normalize
import banglanltk as bn
import copy

filepath = '/content/train.txt'
output_file = '/content/validation.jsonl'

with open(filepath) as fp:
    lines = fp.readlines()
    cnt = 1
    all_data = []
    tokens = []
    tags = []
    for line in lines:
        
        # print("Line {}: {}".format(cnt, line.strip()))
        line = line.strip()
        if line == "":
            all_data.append({'tokens': tokens, 'tags': tags})
            # print(tokens, tags)
            tokens = []
            tags = []
            continue
        words = line.split(' _ _ ')
        # print(words)
        if len(words[0]) == 1:
            token = words[0]
        else:
            token = normalized_text = normalize(
                words[0],
                unicode_norm="NFKC",          # type of unicode normalization (default "NFKC")
                punct_replacement="",       # an optional string or callable for replacing the punctuations (default `None`, i.e. no replacement)
                url_replacement="",         # an optional string or callable for replacing the URLS (default `None`, i.e. no replacement)
                emoji_replacement="",       # an optional string or callable for replacing the emojis (default `None`, i.e. no replacement)
                apply_unicode_norm_last=True  # whether to apply the unicode normalization before or after rule based replacements (default True)        
            )
        if token == "":
            continue
        tokens.append(token)
        tags.append(words[1])
        cnt += 1
```

```python
# read word from json file
import json
from normalizer import normalize
from collections import OrderedDict
import banglanltk as bn
import copy

filepath = '/content/train.jsonl'

# list to contain all the data from json file

all_data = []
# read from a json file and build a list of dictionaries

with open(filepath, 'r', encoding='utf-8') as json_file:
    for line in json_file:
        all_data.append(json.loads(line))
```

```python
# to remove any specific pattern form the token
def removeTokenWithPat(all_data, pat):
  corrected_all_data = []
  for data in all_data:
    tokens = []
    tags = []
    for i in range(len(data["tokens"])):
      if data["tokens"][i] == pat:
        continue
      tokens.append(data["tokens"][i])
      tags.append(data["tags"][i])
    corrected_all_data.append({'tokens': tokens, 'tags': tags})
  
  return corrected_all_data
```

```python
# Testing the length of removed data
len(removeTokenWithPat(all_data, ""))
```

```python
# Random Word Swap
def performSwapWords(data):
  # print(data, len(data["tokens"]))
  prob = random.uniform(0.3, 0.8)
  words = int(len(data["tokens"])*prob)
  if words==0:
    return data
  random_indices = [random.randint(0, len(data["tokens"])-1) for _ in range(words)]
  for idx in random_indices:
    # print(idx)
    token, tag = data["tokens"][idx], data["tags"][idx]
    temp_token, temp_tag = token, tag
    # print(data["tokens"][idx], data["tokens"][(idx+1)%len(data["tokens"])])
    if data["tags"][idx][0] == "B" and data["tags"][(idx+1)%len(data["tags"])][0] == "I":
      data["tokens"][idx] = data["tokens"][(idx+1)%len(data["tokens"])] 
      data["tokens"][(idx+1)%len(data["tokens"])] = temp_token

    elif data["tags"][idx][0] == "I" and data["tags"][(idx+1)%len(data["tags"])][0] == "I":
      data["tokens"][idx] = data["tokens"][(idx+1)%len(data["tokens"])] 
      data["tokens"][(idx+1)%len(data["tokens"])] = temp_token

    elif data["tags"][idx][0] == "O" and data["tags"][(idx+1)%len(data["tags"])][0] == "O":
      data["tokens"][idx] = data["tokens"][(idx+1)%len(data["tokens"])] 
      data["tokens"][(idx+1)%len(data["tokens"])] = temp_token
    # data["tags"][idx] = data["tags"][(idx+1)%len(data["tags"])]
    
    # data["tags"][(idx+1)%len(data["tags"])] = temp_tag

  return data
```

Code to Return all the category wise classification

```python
def return_category_list(all_data):
  category_class = {}
  for data in all_data:
      for i in range(len(data['tokens'])):
          if data['tags'][i] not in category_class:
              category_class[data['tags'][i]] = [data['tokens'][i]]
          else:
              category_class[data['tags'][i]].append(data['tokens'][i])

  return category_class
```

```python
# label Wise token replacement
def LWTR(data, category_class):
  prob = random.uniform(0.3, 0.8)
  words = int(len(data["tokens"])*prob)

  if words==0:
    return data
  random_indices = [random.randint(0, len(data["tokens"])-1) for _ in range(words)]
  data_aug = []
  for idx in random_indices:
    num_times = random.randint(1, 5)
    for i in range(num_times):
      replace_idx = random.randint(0, len(category_class[data["tags"][idx]])-1)
      # print(replace_idx)
      data["tokens"][idx] = category_class[data["tags"][idx]][replace_idx]
      d = copy.deepcopy(data)
      data_aug.append(d)
  return data_aug
```

```python

def replaceWithSynonyms(data):
  prob = random.uniform(0.2, 0.6)
  words = int(len(data["tokens"])*prob)

  if words==0:
    return []
  aug_data = []
  random_indices = [random.randint(0, len(data["tokens"])-1) for _ in range(words)]
  for idx in random_indices:
    if len(data["tokens"][idx]) == 1:
      continue
    syn_words = bn.synonym(data["tokens"][idx])
    if len(syn_words) >3:
      syn_words = syn_words[:3]
    # print("chosen word:", data["tokens"][idx])
    # print("syn: ", syn_words)
    for s_word in syn_words:
      if s_word == data["tokens"][idx]:
        continue
      d = copy.deepcopy(data)
      
      d["tokens"][idx] = s_word
      # print(s_word, data["tokens"][idx])
      aug_data.append(d.copy())
  return aug_data
```

```python
non_other_tags = ["B-CORP", "B-CW", "B-GRP", "B-LOC", "B-PER", "B-PROD", "I-CORP", "I-CW", "I-GRP", "I-LOC", "I-PER", "I-PROD"]
```

```python
def restructureTags(tags):
  for i in range(0, len(tags)-1):
    if tags[i][0] == "B":
      if tags[i+1][0] == "I" and tags[i][2:]!=tags[i+1][2:]:
        tags[i+1] = "B-" + tags[i+1][2:]
      if tags[i+1][0] == "B" and tags[i][2:]==tags[i+1][2:]:
        tags[i+1] = "I-" + tags[i+1][2:]
    elif tags[i][0] == "O":
      if tags[i+1][0] == "I":
        tags[i+1] = "B"+tags[i+1][1:]
    elif tags[i][0] == "I":
      if tags[i+1][0] == "B" and tags[i+1][2:]==tags[i][2:]:
        tags[i+1] = "I" + tags[i+1][1:]
      if tags[i+1][0]=="I" and tags[i+1][2:]!=tags[i][2:]:
        tags[i+1] = "B" + tags[i+1][1:]
  return tags

def randomInsertion(data, category_class):
  num_loops = random.randint(1, 3)
  aug_data = []
  for _ in range(num_loops):
    idx_count = 0;
    tokens = []
    tags = []
    while idx_count < len(data["tokens"]):
        if (random.uniform(0, 1) < 0.2):
          tag = non_other_tags[random.randint(0, len(non_other_tags)-1)]
          token = category_class[tag][random.randint(0, len(category_class[tag])-1)]
          
          tokens.append(token)
          tags.append(tag)
        else:
          tokens.append(data["tokens"][idx_count])
          tags.append(data["tags"][idx_count])
          idx_count += 1
    tags = restructureTags(tags)
    aug_data.append({"tokens": tokens, "tags": tags})
  return aug_data 
```

```python
def randomDeletion(data, category_class):
  num_loops = random.randint(1, 3)
  aug_data = []
  for _ in range(num_loops):
    idx_count = 0;
    tokens = []
    tags = []
    while idx_count < len(data["tokens"]):
        if (random.uniform(0, 1) < 0.5) and data["tags"][idx_count]=="O":
          idx_count+=1
        else:
          tokens.append(data["tokens"][idx_count])
          tags.append(data["tags"][idx_count])
          idx_count += 1
    aug_data.append({"tokens": tokens, "tags": tags})
  return aug_data 
```

```python
print(all_data[2])
d = randomDeletion(all_data[2], category_class)
for s in d:
  print(s)
```

```python
aug_data = []
def augmentData(all_data):
  # all_data = removeTokenWithPat(all_data, "")
  category_class = return_category_list(all_data)
  for data in all_data:
    for _ in range(2):
      d1 = performSwapWords(data)
      d2 = LWTR(data, category_class)
      aug_data.append(d1)
      for dl in d2:
        aug_data.append(dl)
    d = replaceWithSynonyms(data)
    for dl in d:
      aug_data.append(dl)
    d = randomInsertion(data, category_class)
    for dl in d:
      aug_data.append(dl)
    d = randomDeletion(data, category_class)
    for dl in d:
      aug_data.append(dl)
  print(len(aug_data), len(all_data))
```

```python
augmentData(all_data)
print(aug_data[:10])
```
